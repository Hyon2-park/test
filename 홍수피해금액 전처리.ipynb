```python
import pandas as pd
import numpy as np

df=pd.read_csv("연도별 피해액.csv",encoding='cp949')
df1=pd.read_csv('경사도.csv',encoding='cp949')
```

```python
df['행정유역명']=df['행정유역명'].str.split().str[1]
```

```python
df.rename(columns={'행정유역명':'행정구역명'},inplace=True)
df
```

```python
#지역별 10년치 피해 누적액
df_grouped=df.groupby('행정구역명')[['이재민 인','건물피해액']].sum().reset_index()
type(df_grouped)
```

```python
#컬럼명 통일
df1.rename(columns={'SIGUNGU_NM': '행정구역명','SIGUNGU_CD':'시군구코드','_mean':'경사도_평균'}, inplace=True)
```

```python
df1
```

```python
#행정구역명으로 병합
reg=pd.merge(df1,df_grouped,on='행정구역명',how='outer')
reg=reg[['행정구역명','경사도_평균','이재민 인','건물피해액']]
reg = reg.rename(columns={'이재민 인': '이재민인'})
```

```python
#선형회귀로 피해액과 경사도의 선형관계 분석
import statsmodels.formula.api as sts
model = sts.ols('건물피해액 ~ 경사도_평균', data=reg).fit()
model.summary()
```

```python
model2 = sts.ols('이재민인 ~ 경사도_평균', data=reg).fit()
model2.summary()
```
