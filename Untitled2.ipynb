```python
import pandas as pd
import numpy as np

df=pd.read_csv("과거 홍수 피해 금액.csv",encoding='cp949')
```

```python
# 행정구역별 데이터만 남기기
df
```

```python
df1=df.loc[(df['년도']>2011) & (df['행정구역코드여부']=='행정구역')].reset_index(drop=True)
df1.drop(columns='주소',inplace=True)
```

```python
filtered_df=df1.groupby(['행정유역명','년도'])[['이재민 인','건물피해액']].sum()
```

```python
filtered_df.to_csv('연도별 피해액.csv',encoding='cp949')
```
